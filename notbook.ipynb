## Analysis

### 1. Find Two Likely Errors in Land Registry Data

### 2. Calculate Sales and Average Price for London Boroughs in 2023

```python
nspl_data.head(5)
```

```python
# Filter data for 2023 3= date of transfer,4=postcode
df['3'] = pd.to_datetime(df['3'])
df_2023 = df[df['3'].dt.year == 2023]

# Merge datasets on postcode
df_merged = df_2023.merge(nspl_data, left_on='4', right_on='pcds')

```

```python
df_merged.head(5)
```

```python
# Calculate number of sales and average price by London Borough 11-london_borough
borough_summary = df_merged.groupby('11').agg({
    'price': ['count', 'mean']
}).reset_index()

borough_summary.columns = ['London Borough', 'Number of Sales', 'Average Sale Price']
print(borough_summary)
# Compare results to historical data to identify anomalies. This provides insights into the number of transactions and average sale prices by borough.
```

### 3. Count New Build Flats Sold in Each UK Region Since 2020

```python
# Filter for new build flats property_type-7,new_build-6,date_of_transfer-3,region-14
df_new_build_flats = df[(df['7'] == 'F') & (df['6'] == 'Y')]
df_new_build_flats_2020 = df_new_build_flats[df_new_build_flats['3'].dt.year >= 2020]

# Count sales by region
region_sales = df_new_build_flats_2020.groupby('14').size()
print(region_sales)
#Analyzes new build flats sold in each region to identify trends. or New builds and flats might have different market dynamics compared to existing properties.
```

### 4. Plot Number of Sales Per Week Since 2020

```python
# Filter data since 2020
df_since_2020 = df[df['date_of_transfer'].dt.year >= 2020]

# Set the date as index
df_since_2020.set_index('date_of_transfer', inplace=True)

# Resample data to weekly frequency
weekly_sales = df_since_2020.resample('W').size()

# Plot the number of sales per week
weekly_sales.plot(figsize=(12, 6))
plt.title('Number of Sales Per Week Since 2020')
plt.xlabel('Week')
plt.ylabel('Number of Sales')
plt.show()

#Trends may reflect seasonality or market events

```
